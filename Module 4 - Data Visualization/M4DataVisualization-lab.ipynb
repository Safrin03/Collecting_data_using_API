## Demo: How to run an sql query


```python
# List all the tables in the database
QUERY = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

# Execute the query and fetch the table names
table_names = conn.execute(QUERY).fetchall()

# Print the list of table names
print(table_names)
```

```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
median_workweekhrs = age_group['WorkWeekHrs'].median()
median_coderevhrs = age_group['CodeRevHrs'].median()

median_data = pd.DataFrame({'WorkWeekHrs': [median_workweekhrs], 'CodeRevHrs': [median_coderevhrs]})

median_data.plot(kind='bar', stacked=True, figsize=(8, 6))
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
median_comp_by_age = filtered_df.groupby('Age')['ConvertedComp'].median()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-')
plt.title("Median ConvertedComp by Age (45 to 60)")
plt.xlabel("Age")
plt.ylabel("Median ConvertedComp")
plt.grid(True)
plt.xticks(median_comp_by_age.index)
plt.tight_layout()
plt.show()
```

```python
age_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
median_comp = age_df.groupby('Age')['ConvertedComp'].median()
plt.plot(median_comp.index, median_comp.values, marker='o', linestyle='-')
plt.title("Correlation between age and median converted comp")
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
branch_counts = df['MainBranch'].value_counts()
plt.figure(figsize=(8, 6))
branch_counts.plot(kind='barhb, color='skyblue')
plt.title("Distribution of MainBranch")
plt.xlabel("Count")
plt.ylabel("MainBranch")
plt.show()
```
